## 0. Setup Inicial (Já preenchido)
Vamos importar as bibliotecas e recriar o nosso `df_base` com a Atividade e o Sono que vimos na aula.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

# Links brutos do GitHub (Simulando o acesso a APIs)
url_act = 'https://raw.githubusercontent.com/jaimecirne/Aulas/main/Activity.csv'
url_sleep = 'https://raw.githubusercontent.com/jaimecirne/Aulas/main/Sleep.csv'

# Ingestão e Limpeza Rápida da Aula
df_act = pd.read_csv(url_act)
df_sleep = pd.read_csv(url_sleep)

act_agrupado = df_act.groupby('Date').agg({'Duration (min)': 'sum'}).reset_index()
sleep_agrupado = df_sleep.groupby('Date').agg({'Deep Sleep (min)': 'sum', 'Light Sleep (min)': 'sum', 'REM Sleep (min)': 'sum'}).reset_index()
sleep_agrupado['Sono Total (horas)'] = (sleep_agrupado['Deep Sleep (min)'] + sleep_agrupado['Light Sleep (min)'] + sleep_agrupado['REM Sleep (min)']) / 60

# O nosso DataFrame Base (Atividade + Sono)
df_base = pd.merge(act_agrupado, sleep_agrupado[['Date', 'Sono Total (horas)']], on='Date', how='outer')

print("Setup concluído! O seu df_base tem as colunas:", df_base.columns.tolist())
```
